```python
import matplotlib.pyplot as plt

# ==== Data ====
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
prices = [48, 50, 48, 60, 62, 59, 47, 45, 45, 42, 40, 37]

# ==== Plot ====
plt.figure(figsize=(10, 5))
plt.plot(months, prices, color='red', linewidth=1.5, marker='o',
         markersize=2, markerfacecolor='white', markeredgecolor='black')

# ==== Add value labels (Mar & Jul pushed down) ====
for i, price in enumerate(prices):
    y_offset = -1.2 if months[i] in ['Mar', 'Jul'] else 0.5
    plt.text(i, price + y_offset, str(price), ha='center', fontsize=8)

# ==== Add minimalist up/down/flat icons ====
for i in range(len(prices) - 1):
    dy = prices[i + 1] - prices[i]
    if dy > 0:
        symbol, y_offset = '', 1.2
    elif dy < 0:
        symbol, y_offset = '', -1.2
    else:
        symbol, y_offset = '', 0.5

    plt.text(i + 0.5, (prices[i] + prices[i + 1]) / 2 + y_offset,
             symbol, color='green', fontsize=10, ha='center', fontweight='bold')

# ==== In-chart caption box ====
caption = "Prices peaked in May at $62, then steadily declined to $37 by Dec."
plt.text(0.5, 0.92, caption, fontsize=8, transform=plt.gca().transAxes,
         bbox=dict(facecolor='white', edgecolor='gray', boxstyle='round,pad=0.5'),
         verticalalignment='top')

# ==== Titles and Labels ====
plt.grid(True, linestyle='--', alpha=0.5)
plt.title('Oil Prices Spike Between April and June')
plt.xlabel('Month')
plt.ylabel('Price (US$)')
plt.tight_layout()

# ==== Save and Show ====
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
prices = [48, 50, 48, 60, 62, 59, 47, 45, 45, 42, 40, 37]

colors = []
for i in range(len(prices)-1):
    if prices[i+1] > prices[i]:
        colors.append('#ff9966')  # sunrise tone
    else:
        colors.append('#6699cc')  # twilight tone

fig, ax = plt.subplots(figsize=(10, 5))
for i in range(len(prices)-1):
    ax.plot(months[i:i+2], prices[i:i+2], color=colors[i], linewidth=2.5)

ax.plot(months, prices, marker='o', markersize=3, markerfacecolor='white', markeredgecolor='black', linestyle='')

for i, price in enumerate(prices):
    y_offset = -1.2 if months[i] in ['Mar', 'Jul'] else 0.5
    ax.text(i, price + y_offset, str(price), ha='center', fontsize=8)

ax.set_title('Mood-Driven Oil Price Chart (Sunrise = Rise, Twilight = Fall)', fontsize=12)
ax.set_xlabel('Month')
ax.set_ylabel('Price (US$)')
ax.grid(True, linestyle='--', alpha=0.4)
plt.tight_layout()
plt.savefig('mood_driven_styling_chart.png', dpi=600)
plt.show()
```

```python
import matplotlib.pyplot as plt

# ==== Data ====
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
prices = [48, 50, 48, 60, 62, 59, 47, 45, 45, 42, 40, 37]

# ==== Caption rules (simple trend summaries) ====
captions = [
    "Price starts at $48",
    "Price increases to $50",
    "Price dips slightly",
    "Sharp jump to $60",
    "Peak at $62",
    "Slight decline to $59",
    "Drops to $47",
    "Continues falling",
    "Stable at $45",
    "Decline resumes",
    "Falls to $40",
    "Lowest at $37"
]

# ==== Plot each month in its own frame ====
fig, axes = plt.subplots(3, 4, figsize=(13, 7))
axes = axes.flatten()

for i in range(12):
    ax = axes[i]

    # Line plot up to current month
    ax.plot(months[:i+1], prices[:i+1], color='black', linewidth=1.5, marker='o',
            markersize=3, markerfacecolor='white', markeredgecolor='black')

    # Add value labels
    for j in range(i+1):
        y_offset = -1.2 if months[j] in ['Mar', 'Jul'] else 0.5
        ax.text(j, prices[j] + y_offset, str(prices[j]), ha='center', fontsize=7)

    # Title = month, Caption = summary
    ax.set_title(months[i], fontsize=9)
    ax.text(0.5, -0.25, captions[i], fontsize=7, ha='center', va='top', transform=ax.transAxes)

    # Style
    ax.set_ylim(30, 70)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.grid(True, linestyle='--', alpha=0.3)

# Overall title
plt.suptitle('Frame-by-Frame Oil Price Timeline with Captions', fontsize=14)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.savefig('frame_by_frame_timeline_with_captions.png', dpi=600)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
prices = [48, 50, 48, 60, 62, 59, 47, 45, 45, 42, 40, 37]

# === Color coding based on trend
colors = []
for i in range(len(prices) - 1):
    if prices[i + 1] > prices[i]:
        colors.append('#ff9966')  # sunrise tone
    else:
        colors.append('#6699cc')  # twilight tone

# === Plot
fig, ax = plt.subplots(figsize=(10, 5))

# === Line segments with mood coloring
for i in range(len(prices) - 1):
    ax.plot(months[i:i+2], prices[i:i+2], color=colors[i], linewidth=2.5)

# === Dots
ax.plot(months, prices, marker='o', markersize=3,
        markerfacecolor='white', markeredgecolor='black', linestyle='')

# === Value labels
for i, price in enumerate(prices):
    y_offset = -1.2 if months[i] in ['Mar', 'Jul'] else 0.5
    ax.text(i, price + y_offset, str(price), ha='center', fontsize=8)

# === Titles and labels
ax.set_title('Mood-Driven Oil Price Chart (Sunrise = Rise, Twilight = Fall)', fontsize=12)
ax.set_xlabel('Month')
ax.set_ylabel('Price (US$)')
ax.grid(True, linestyle='--', alpha=0.4)

# === Add caption below chart
caption = "Oil prices peaked in May at $62 after a steep rise from March, followed by a steady decline to $37 by December."
fig.text(0.5, -0.05, caption, wrap=True, ha='center', fontsize=9)

plt.tight_layout(rect=[0, 0.03, 1, 1])  # Make space for the caption
plt.savefig('mood_driven_styling_chart_with_caption.png', dpi=600)
plt.show()
```
